# 15장 확률적인 설명변수

## 임금방정식에서 IQ의 통제

```python
import pandas as pd
import statsmodels.formula.api as smf

wage2 = pd.read_csv('csv/wooldridge/wage2.csv')
fm = 'lwage~educ+exper+tenure+married+south+urban+black'
smf.ols(fm, data=wage2).fit().summary(slim=True)
```

```python
smf.ols(fm+'+IQ', data=wage2).fit().summary(slim=True)
```
